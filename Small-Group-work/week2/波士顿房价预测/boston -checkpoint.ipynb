```python
# CRIM：城镇的人均犯罪率
# ZN：大于25,000平方英尺的地块的住宅用地比例。
# INDUS：每个镇的非零售业务英亩的比例。
# CHAS：查尔斯河虚拟变量（如果环河，则等于1；否则等于0）
# NOX：一氧化氮的浓度（百万分之几）
# RM：每个住宅的平均房间数
# AGE：1940年之前建造的自有住房的比例
# DIS：到五个波士顿就业中心的加权距离
# RAD：径向公路通达性的指标
# TAX：每$ 10,000的全值财产税率
# PTRATIO：各镇的师生比率
# B：计算方法为1000（Bk-0.63）²，其中Bk是按城镇划分的非裔美国人的比例
# LSTAT：底层人口的百分比
# MEDV：自有住房数的中位数，单位为$ 1000
```

```python

# 一.导入处理数据集和三大库
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv('boston_housing_data.csv'  )
df
```

```python
# 1.1检查缺失值
missing_value=df.isnull().sum()#统计每列数据的缺失值
missing_value
#发现房价缺失值为54
```

```python
# 1.2利用每列均值弥补缺失值
mean_fill=df.fillna(df.mean())
df=mean_fill
df
```

```python
#  2.异常值处理
```

```python
#2.1将数据集每个特征名放入列表中
sr1=pd.DataFrame(df,columns=df.columns)#得到每列的特征名称
sr1
column_names=list(sr1.columns)
column_names
```

```python
#2.2删除异常值
outlier_threshold=3 #设置阈值

#for循环遍历每列的数据
for column_name in column_names:
    mean=sr1[column_name].mean()# 计算均值
    std=sr1[column_name].std()  #计算标准差
#计算实际值与均值的偏差，并于阈值*标准差作比较，小于的为正常值
    df=sr1[abs(sr1[column_name]-mean)<outlier_threshold*std]
df
#得到删除异常值的数据（发现有17个异常数据样本）
```

```python

df2=pd.concat([df1['LSTAT'],df1['INDUS'],df1['CRIM'],df1['MEDV']],axis=1)
df2
```

```python
# 3.创建训练集和测试集
```

```python
#划分特征和目标变量
x=df2[['LSTAT']].values  #通过.values转化为数组
y=df2['MEDV'].values

#设置随机种子
np.random.seed(42)

#随机打乱数据
idx=np.arange(x.shape[0])  #创建一个索引数据样本下标的数组
np.random.shuffle(idx)  #打乱下标
x=x[idx]
y=y[idx]

#划分训练集和测试集
split_ratio=0.8
split_index=int(split_ratio*x.shape[0])  #返回值为80

 #得到80%训练集和20%测试集
x_train,x_test=x[:split_index],x[split_index:]
y_train,y_test=y[:split_index],y[split_index:]
y_train
```

```python
#三.构建线性回归模型
#回归模型：y=theta_0+theta_1*LSTAT+theta_2*RM
```

```python
#1.梯度下降法创建
```

```python

# 添加偏置项
X_train = np.c_[np.ones(x_train.shape[0]), x_train]
X_train
```

```python
#设置学习率
alpha=0.001
#样本数
m=489
```

```python
#1.2定义代价函数
def cost_function(theta,X,Y):
    diff=np.dot(X,theta)-Y    
    return (1/(2*m)) *np.dot(diff.transpose(), diff)
```

```python
#1.3定义代价函数对应的梯度函数
def gradient_function(theta,X,Y):
    diff=np.dot(X,theta)-Y
    return (1/m)*np.dot(X.T,diff)
```

```python
#1.4梯度下降迭代
def gradient_descent(X,Y,alpha):
    #初始化theta_0theta_1
    global theta
    theta=np.random.rand(2,)
    gradient=gradient_function(theta,X,Y)
    for i in range(500000):
        theta=theta-alpha*gradient
        gradient=gradient_function(theta,X,Y)
    return theta
```

```python
# 得到最优参数向量
optimal=gradient_descent(X_train,y_train,alpha)
optimal
```

```python
#均方误差
print(cost_function(optimal,X_train,y_train))
```

```python
#得出线性回归图
plt.figure(figsize=(10,5))
plt.title("线性回归预测-梯度下降法")
plt.xlabel("人口中地位较低人群的百分比")
plt.ylabel("房价")
x1=X_train
y1=optimal[0]+optimal[1]*x1
plt.scatter(x1, y1, s=30, c="red", marker="s", label='拟合值')  #拟合值
plt.scatter(x_train, y_train, c="blue", marker="s", label='真实值', alpha=0.5)
plt.scatter(x_test, y_test, c="gold", marker="s", label='测试集数据')
plt.plot(x1, y1, c='red')  # 绘制拟合直线

plt.legend()
plt.show()
```

```python
#2.1最小二乘法
#得到线性回归系数和阙值
w_least_squares = np.linalg.inv(X_train.T.dot(X_train)).dot(X_train.T).dot(y_train)
w_least_squares 
```
